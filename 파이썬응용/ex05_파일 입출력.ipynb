```python
f = open("./smhrd/bbb.txt",'r',encoding='utf-8')
data = f.readline()
f.close()

print(data)
```

```python
preprocessing_data = []
for i in range(len(data)):
    preprocessing_data.append(data[i].replace("\n","").split("\t")[1])
    
preprocessing_data
```

###### 예제1 

```python
#특정 키워드를 가진 리뷰만 골라내는 findReview 함수
def findReview(data,keyword):
    review=[]
    for r in data:
        if keyword in r:
            review.append(r)
    return review
```

```python
findReview(preprocessing_data,"꿀잼")
```
