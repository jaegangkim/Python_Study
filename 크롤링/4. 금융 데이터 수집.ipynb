#### 100등까지의 종목, 현재가, 거래량  수집 ,   csv파일로 제작
- https://finance.naver.com/sise/sise_quant.naver?sosok=0

```python
import requests as req
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
res = req.get("https://finance.naver.com/sise/sise_quant.naver?sosok=0")
soup = bs(res.text,'lxml')
name_list=[i.text for i in soup.select("a.tltle")]
```

```python
price = soup.select("table.type_2 tr td.number:nth-child(3)")
price_list=[]
for i in price:
    price_list.append(i.text)
len(price)
```

```python
deal = soup.select("table.type_2 tr td.number:nth-child(6)")
deal_list=[]
for i in deal:
    deal_list.append(i.text)
```

```python
dic = {"제목":name_list,"현재가":price_list,"거래량":deal_list}
df = pd.DataFrame(dic)
df.index=range(1,101)
df.index.name = "N"
df
```

```python
df.to_csv("./stock.csv",encoding="euc-kr")
```
