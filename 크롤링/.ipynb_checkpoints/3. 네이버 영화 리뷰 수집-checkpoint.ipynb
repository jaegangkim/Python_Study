```python
import requests as req
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
res = req.get("https://movie.naver.com/movie/bi/mi/point.naver?code=81888")
res
```

```python
soup = bs(res.text, 'lxml')
```

```python
# 관람객 평점 : 눈에 보이지만 수집 할 수 없음.
# 해당 태그는 iframe태그로 만들어져 있음
data = soup.select("div.score_reple > p")
data
```

iframe : 다른 주소에 있는 정보를 보여 줄 때 사용  
    찾아서 src속성에 주소가 있는데 메인서버 주소가 빠져있음  
    https://movie.naver.com/ 등을 입력 해야 함

```python
# iframe에 숨겨진 링크를 찾아서 다시 연결
res = req.get("https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=81888&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false")
soup = bs(res.text,'lxml')
review = soup.select("div.score_reple > p > span:nth-child(2)")
# nth-child(2) : 두번째 span만 가져오기 
```

```python
for i in review:
    print(i.text.strip())    # \r\n\t 등 공백 제거 후 프린트
```

```python
# last_child 해도 되긴 할 듯?

# 2번째 span에 리뷰가 없는 경우도 있음.
# span전체를끌어와서 관람객이라는 글자들만 삭제하기.
review = soup.select("div.score_reple > p > span")
ico = soup.select('span.ico_viewer')   # 관람객 있는 태그
for i in ico:
    i.extract()
```

```python
for i in review:
    print(i.text.strip())
```

10 페이지까지 출력하기   
  
반복문의 진행상황을 체크 할 수 있는 라이브러리(로딩바 느낌)

```python
from tqdm import tqdm_notebook
```

```python
review = []
for i in tqdm_notebook(range(1,11)):
    res = req.get(f"https://movie.naver.com/movie/bi/mi/pointWriteFormList.naver?code=81888&type=after&isActualPointWriteExecute=false&isMileageSubscriptionAlready=false&isMileageSubscriptionReject=false&page={i}")
    soup = bs(res.text,'lxml')
    ico = soup.select('span.ico_viewer')   # 관람객 있는 태그
    for j in ico:
        j.extract()
    review += soup.select("div.score_reple > p > span")
    
for i in review:
        print(i.text.strip())
```
