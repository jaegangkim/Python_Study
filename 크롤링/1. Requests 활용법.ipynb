requests : 브라우저의 역할을 대신해줌
- 사이트의 요청과 응답정보를 처리

```python
import requests as req
```

```python
# 1. 네이버 url을 통해서 사이트 요청
res = req.get('http://www.naver.com')
```

```python
# 2. 응답하는 페이지에서 원하는 정보 추출
res.text  # html 코드를 문자열로 가져옴
```

#### BeautifulSoup 
- 3. 가지고 온 문자열을 html코드로 변경

```python
from bs4 import BeautifulSoup as bs
```

```python
# 문자열을 코드로 변경 , lxml
soup = bs(res.text, "lxml")
```

```python
# a태그를 가지고 있는 값 추출
soup.select('a')
```

```python
# 선택자, 태그를 사용해서 더 자세하게 추출, 리스트로 저장
data = soup.select('a.nav')
```

```python
for i in data:
    print(i.text)
```
